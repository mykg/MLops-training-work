```python
def fol(i,color):
    folium.Marker(location=mark, 
                  popup=suburb[i],
                 icon=folium.Icon(color=color, icon_color='red', icon='tint')).add_to(newmap)
```

```python
i = 0
for mark in places:
    if dataset['Number of single-residential properties_number'][i] >= 1000:
        fol(i,'white')
        
    else:
        fol(i,'green')
        
    i+=1
```
