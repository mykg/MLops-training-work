```python
from scipy import misc
```

```python
img = misc.ascent()
```

```python
img.shape
```

```python
plt.imshow(img, cmap='gray')
```

```python
kernel = np.array([
        [1,2,1],
        [0,0,0],
        [-1,-2,-1] ])
```

```python
plt.imshow(kernel, cmap='gray')
```

```python
from scipy.signal import convolve2d
```

```python
cimg = convolve2d(img, kernel)
```

```python
cimg.shape
```

```python
plt.imshow(cimg, cmap='gray')
```
