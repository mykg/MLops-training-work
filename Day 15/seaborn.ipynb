```python
import seaborn as sns
sns.set() # give some look 
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.columns
```

```python
bill = tips['total_bill']
```

```python
bill.max()
```

```python
bill.min()
```

```python
sns.distplot(bill, kde=False)
```

```python
sns.distplot(bill, kde=False, bins=50)
```

```python
# univariate we use distplot

# bivariate we can use this
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```

```python
sns.pairplot(data=tips)
```

```python
# to differentiate 
sns.pairplot(data=tips, hue='sex')
```

```python
sns.pairplot(data=tips, hue='sex', palette='cool')
```

```python
sns.barplot(data=tips, x='sex', y='total_bill') # creates bar graph
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', palette='rainbow')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', palette='rainbow', hue='sex') # when we want per something
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', palette='rainbow', hue='sex')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', palette='Set1', hue='sex')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', palette='Set1', hue='sex', jitter=False)
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', palette='Set1', hue='sex', dodge=True, jitter=False)
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', palette='Set1', hue='sex', dodge=True)
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(), cmap='coolwarm', annot=True)
```
